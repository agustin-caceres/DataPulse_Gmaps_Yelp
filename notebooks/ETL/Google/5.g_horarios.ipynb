En este notebook armaré una tabla que contenga los dias y horarios en los que abre cada negocio de los estados y categorías seleccionados.

```python
import pandas as pd
import os
```

```python
# Leemos la tabla v1.g_sitios que tenemos hasta el momento.
g_sitios = pd.read_parquet(r"C:\Users\Juli\Desktop\Data Science\Archivos Proyecto Final\v1.g_sitios.parquet", engine='pyarrow')
g_sitios.info()
```

```python
g_sitios
```

Cambiaré el nombre de la columna "gmap_id" por "id_negocio".

```python
g_sitios = g_sitios.rename(columns={'gmap_id': 'id_negocio'})
```

```python
# Creamos la tabla horarios
g_horarios = g_sitios[['id_negocio','monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday','sunday']]
g_horarios
```

```python
g_horarios
```

```python
g_horarios.info()
```

Verificamos si hay nulos.

```python
nulos = g_horarios.isnull().sum()
nulos
```

Guardamos el dataframe en un archivo parquet para luego cargar a nuestra base de datos.

```python
g_horarios.to_parquet(r"C:\Users\Juli\Desktop\Data Science\Archivos Proyecto Final\v0_horarios.parquet")
```

```python
# Eliminamos los dias de g_sitios
g_sitios = g_sitios.drop(columns=['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'])

# Guardamos el nuevo Dataframe.
os.makedirs('../../../data', exist_ok=True)
g_sitios.to_parquet('../../../data/google_sitios_v1.parquet', index=False)

# Muestra parcial de la tabla:
g_sitios.head()
```
