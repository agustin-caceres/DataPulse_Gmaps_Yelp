Importando Librerias

```python
import pandas as pd
```

Leyendo el Datasets

```python
yelp_user = pd.read_parquet("datasets/yelp/user.parquet")
```

Como vienen los Datos

```python
yelp_user.head()
```

Verificación de Nulos

```python
for column in yelp_user.columns:
    print(f"Columna {column}: {yelp_user[column].isnull().sum()}")
```

Verificando si hay filas repetidas

```python
print(f"Total Filas como viene la data {yelp_user.shape[0]}")

print(f"Total filas Unicas {yelp_user.unique().shape[0]}")
```

```python
yelp_user = yelp_user.unique()

yelp_user.shape
```

Creando la Tabla `yelp_friends_user`

```python
yelp_friends_user = yelp_user[['user_id', 'friends']]

yelp_friends_user.columns = ['user_id', 'friend_id']
```

Creando la Tabla `yelp_elite_user`

```python
yelp_elite_user = yelp_user[['user_id', 'elite']]
```

Ahora puedo Borrar las columnas `friends` y `elite` de la tabla `yelp_user`, luego vamos a guardar esta tabla y por ultimo borraremos la variable para liberar memoria

```python
columnas_a_eliminar = ['friends', 'elite']
yelp_user = yelp_user.drop(columns=columnas_a_eliminar)

yelp_user.to_parquet('datasets/yelp/yelp_user.parquet')
```

Toca trabajar la tabla `yelp_friends_user`

```python
print(yelp_friends_user.shape)
yelp_friends_user.head()
```

Eliminando los None

```python
mask = yelp_friends_user['friend_id'] == 'None'

yelp_friends_user = yelp_friends_user.loc[~mask]

print(yelp_friends_user.shape)
yelp_friends_user.head()
```

Se crea una columna nueva que sea de tipo LISTA a partir de `friend_id` y luego borrare esta ultima

```python
yelp_friends_user['lista_friends'] = yelp_friends_user['friend_id'].apply(lambda x: x.split(','))
```

Se borra la columna `friend_id`

```python
yelp_friends_user = yelp_friends_user.drop(columns=['friend_id'])

print(yelp_friends_user.shape)
yelp_friends_user.head()
```

Se utiliza el metodo `explode` de los dataframes y luego se cambian los nombres de las columnas

```python
yelp_friends_user = yelp_friends_user.explode('lista_friends')

yelp_friends_user.columns = ['user_id', 'friend_id']

print(yelp_friends_user.shape)
yelp_friends_user.head()
```

Guardando la Tabla

```python
yelp_friends_user.to_parquet('datasets/yelp/yelp_friends_user.parquet')
```

Ahora se hara lo mismo con la tabla `yelp_elite_user`

```python
yelp_elite_user.head(50)
```

La mayoria de los usuario no tienen Años Elite... Hay que borrar los registros vacios

```python
mask = yelp_elite_user['elite'] == ''

yelp_elite_user = yelp_elite_user.loc[~mask]

print(yelp_elite_user.shape)
yelp_elite_user.head()
```

Toca crear una columna nueva que sea de tipo Lista a partir del String que hay en la columna elite

```python
yelp_elite_user['year'] = yelp_elite_user['elite'].apply(lambda x: x.split(','))
```

Borrando la columna `elite`

```python
yelp_elite_user = yelp_elite_user.drop(columns=['elite'])
```

Aplicando el metodo `explode` a la columna `year` que hasta ahora es de tipo Lista

```python
yelp_elite_user = yelp_elite_user.explode('year')

yelp_elite_user.head()
```

Eliminando repetidos y Sustituyendo los 20 por 2020

```python
yelp_elite_user = yelp_elite_user.unique()

yelp_elite_user.loc[yelp_elite_user['year'] == '20', 'year'] = '2020'

yelp_elite_user.head()
```

Guardando la Tabla `yelp_elite_user`

```python
yelp_elite_user.to_parquet('datasets/yelp/yelp_elite_user.parquet')
```
